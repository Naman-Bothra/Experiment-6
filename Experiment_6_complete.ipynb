```python
#5 To check whether the given year is leap year or not

year = int(input("Enter the year : "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
  print("The year is leap year")
else:
  print("The year is not a leap year")
```

```python
#7 write a python program to calculate the gross salary of a employee based on
# is basic salary is 10000: HRA is 20%, DA is 80%
# if basic salary is 20000: HRA is 25%, DA is 90%
# Otherwise : HRS is 30%, DA is 90%

salary = int(input("Enter your salary: "))

if salary <= 10000:
  print(f"Your gross salary is: {salary + (salary * 0.2) + (salary * 0.8)}")
elif salary <= 20000:
  print(f"Your gross salary is: {salary + (salary * 0.25) + (salary * 0.9)}")
else:
  print(f"Your gross salary is: {salary + (salary * 0.3) + (salary * 0.9)}")
```

```python
#8 write a python program to calculate the income tax based on
# upto 2,50,000: No tax
# 2,50,001 to 5,00,000: 5%
# 5,00,001 to 10,00,000: 20%
# above 10,00,,000: 30%

inc = int(input("Enter your income: "))

if inc <= 250000:
  print(f"Your tax is: {inc * 0}")
elif inc <= 500000:
  print(f"Your tax is: { (inc - 250000) * 0.05}")
elif inc <= 1000000:
  print(f"Your tax is: {(inc - 500000) * 0.2 + (250000 * 0.05)}")
else:
  print(f"Your tax is: {(250000 * 0.05) + (500000 * 0.2) + (inc - 1000000) * 0.3}")
```

```python
#9 write python program to identify if the given program is vowel or consonant

chr = str(input("Enter a character: "))
vow = ['a','e','i','o','u','A','E','I','O','U']

if chr in vow:
  print(f"{chr} is vowel")
else:
  print("It is a consonant")
```
